```python
1 # Ex : 3차원 배열 슬라이싱
import numpy as np

4 # 데이터 생성 : 60개 원소로 구성된 (3,4,5) 형태의 3차원 배열 생성 (3층, 4행, 5열)
tensor = np.arange(60).reshape((3,4,5))
print (tensor)
print ("="*100)

9 # (1) 특정 층 전체 가져오기 (첫번째 층)
print(tensor[0, :, :]) # tensor[0] 동일-> Shape : (4,5), size = 20
print ("="*100)
12
13 # (2) 모든 층에서 특정 '행'만 가져오기 (모든 층의 첫 번째 행)
print (tensor [ : , 0, : ])  # shape : (3,5),size = 15
print ("="*100)
16# (3) 모든 층에서 왼쪽 벽면( 첫 번째 열)만 수직으로 잘라내기
print (tensor[:,:,0]) # shape : (3,4), size = 12

19 # (4) 정가운데 부분만 즉, 큐브형태로 도려내기
print(tensor[1:3, 1:3, 1:4]) # 1~2층의 1~2행에서 1~3멸 , shape:(2,2,3)
```

# 14. 다차원 배열 프로그래밍
- 벡터화 연산 (데이터병렬 처리)
  - 배열 간 연산
  - 리덕션
    - np.sum(arr, axis=)
    - np.mean(arr, axis=)
    - arr.sum(axis=0)
    - arr.sum(axis=1)
    - arr.sum(axis=(0,1))
  - 조건 리덕션 원리 (Boolean to Integer)
      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

      - np.mean(condition) : 조건이 참(True)인 비율을 구함 (예: 0.7이면 70%가 참)
      - 이 리덕션 기법들을 활용해서 실제 데이터셋에서 "이상치(Outlier)"를 찾기
